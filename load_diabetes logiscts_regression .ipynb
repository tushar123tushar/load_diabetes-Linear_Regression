```python
from sklearn.datasets import load_diabetes
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
load_Data = load_diabetes()
load_Data
```

```python
new_set = pd.DataFrame(load_Data.data)
```

```python
x = new_set
y = load_Data.target
```

```python
x
```

```python
y
```

```python
sns.heatmap(new_set.isnull() , yticklabels=False , cbar=False)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    x, y, test_size=0.33, random_state=42)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

logistc = LogisticRegression()
logistc
```

```python
Parametrs = {'penalty': ['elasticnet', 'l2', 'l1'] , 'C' : [1,2,3,4,5,6,10,5,450,102] , 'max_iter':[100,200,300]}
Gride = GridSearchCV(logistc , param_grid=Parametrs , scoring='accuracy' , cv=5)
```

```python
Gride.fit(X_train , y_train)
```

```python
print(Gride.best_params_)
```

```python
print(Gride.best_score_)
```

```python
##predict
x_pred = Gride.predict(X_test)
```

```python
from sklearn.metrics import classification_report , accuracy_score
```

```python
accuracy_score(x_pred , y_test)
```

```python
classification_report(x_pred , y_test)
```
